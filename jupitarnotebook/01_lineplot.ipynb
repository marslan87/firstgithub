## adding title

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

flow = sns.load_dataset("iris")
flow

# draw a line plot

sns.lineplot(x="sepal_length",y="sepal_width",data=flow)
plt.title("flowers chart")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

flow = sns.load_dataset("iris")
flow

# draw a line plot

sns.lineplot(x="sepal_length",y="sepal_width",data=flow)
plt.title("flowers chart")
plt.xlim(2)
plt.ylim(0)
plt.show()
```
